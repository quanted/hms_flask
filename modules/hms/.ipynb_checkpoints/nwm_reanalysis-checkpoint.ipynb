```python

url = 's3://noaa-nwm-retro-v2-zarr-pds'
var = 'streamflow'
sig_figs = 3
# raw_bounds = nwm.geometry.bounds
# bounds = [round(float(raw_bounds["miny"]), sig_figs), round(float(raw_bounds["maxy"]), sig_figs),
#           round(float(raw_bounds["minx"]), sig_figs), round(float(raw_bounds["maxx"]), sig_figs)
#           ]
```

```python
%%time
ds = xr.open_zarr(fsspec.get_mapper(url, anon=True), consolidated=True)
ds
```

```python
%%time

start_date = "2010-01-01"
end_date = "2010-12-31"
comid = 6177558
nwm = NWM(start_date=start_date, end_date=end_date, comid=comid)
comids = [6177558,6140826,6162981,9345058,8479268,724969,4594599]
# nwm.get_geometry()
# nwm.geometry.bounds
```

```python
%%time
a_comids = []
for c in comids:
    try:
        test = ds["streamflow"].sel(feature_id=c).sel(time=slice("2010-01-01", "2010-01-01"))
        a_comids.append(c)
    except KeyError:
        print(f"COMID: {c} not found")
comids = a_comids
comids
```

```python
%%time
# bounds = nwm.geometry.bounds
# idx = (ds.latitude > round(bounds["miny"],3)) & (ds.latitude < round(bounds["maxy"],3)) & (
#                     ds.longitude > round(bounds["minx"],3)) & (ds.longitude < round(bounds["maxx"],3))
nwm.end_date = nwm.end_date + datetime.timedelta(days=1)
# variables = ["elevation", "order", "qBtmVertRunoff", "qBucket", "qSfcLatRunoff", "q_lateral", "streamflow", "velocity"]
variables = ["streamflow"]
with dask.config.set(**{'array.slicing.split_large_chunks': False}):
    ds_streamflow = ds[variables].sel(feature_id=comids).sel(time=slice(
        f"{nwm.start_date.year}-{nwm.start_date.month}-{nwm.start_date.day}",
        f"{nwm.end_date.year}-{nwm.end_date.month}-{nwm.end_date.day}"
    )).compute()
#     ds_streamflow = ds[variables].isel(feature_id=idx)
#     ds_streamflow = ds_streamflow.sel(time=slice(
#         f"{nwm.start_date.year}-{nwm.start_date.month}-{nwm.start_date.day} 00:00",
#         f"{nwm.end_date.year}-{nwm.end_date.month}-{nwm.end_date.day} 00:00"
#     )).sel(feature_id=comid).compute()

ds_streamflow
```
